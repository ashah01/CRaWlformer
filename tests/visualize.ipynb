```python
import re
from operator import itemgetter
import matplotlib.pyplot as plt
```

### Crawlformer plot

```python
with open('crawlformer.txt', 'r') as f:
    lines = f.readlines()
train = []
valid = []
for out in lines:
    rg1 = re.search('^Epoch \d{1,3}', out)
    if not rg1:
        continue
    losses = re.findall("0.\d{4}", out)
    train.append(float(losses[1]))
    valid.append(float(losses[2]))
```

```python
plt.plot([x for x in range(len(train))], train, label='Train')
plt.plot([x for x in range(len(valid))], valid, label="Valid")
plt.legend()
plt.title("Crawlformer")
plt.show()
```

### GraphGPS plot

```python
with open('graphgps.txt', 'r') as f:
    lines = f.readlines()
train = []
valid = []
for out in lines:
    if out.startswith('train: '):
        train.append(float(re.search("'loss': \d\.\d+", out).group().split()[1]))
    
    elif out.startswith('val: '):
        valid.append(float(re.search("'loss': \d\.\d+", out).group().split()[1]))
```

```python
plt.plot([x for x in range(len(train[:220]))], train[:220], label='Train')
plt.plot([x for x in range(len(valid[:220]))], valid[:220], label="Valid")
plt.legend()
plt.title("GPS")
plt.show()
```
