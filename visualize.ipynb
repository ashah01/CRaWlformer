```python
import re
from operator import itemgetter
import matplotlib.pyplot as plt
```

### Crawlformer plot

```python
with open('crawlformer.txt', 'r') as f:
    lines = f.readlines()
train = []
valid = []
for out in lines:
    rg1 = re.search('^Epoch \d{1,3}', out)
    if not rg1:
        continue
    losses = re.findall("0.\d{4}", out)
    train.append(float(losses[1]))
    valid.append(float(losses[2]))
```

```python
plt.plot([x for x in range(len(train))], train)
```

```python
plt.plot([x for x in range(len(valid))], valid)
```

```python
(valid[230] + valid[231] + valid[232]) / 3
```

```python
(train[230] + train[231] + train[232]) / 3
```

### GraphGPS plot

```python
with open('graphgps.txt', 'r') as f:
    lines = f.readlines()
train = []
valid = []
for out in lines:
    if out.startswith('train: '):
        train.append(float(re.search("'loss': \d\.\d+", out).group().split()[1]))
    
    elif out.startswith('val: '):
        valid.append(float(re.search("'loss': \d\.\d+", out).group().split()[1]))
```

```python
plt.plot([x for x in range(len(train))], train)
```

```python
plt.plot([x for x in range(len(valid))], valid)
```

```python
(valid[230] + valid[231] + valid[232]) / 3
```

```python
(train[230] + train[231] + train[232]) / 3
```
